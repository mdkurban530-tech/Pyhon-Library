# Matplotlib Example workflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter notebook 
%matplotlib inline
import matplotlib.pyplot as plt 
# 1. Prepare data
x=[1,2,3,4] 
y=[11,22,33,44] 
# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,5)) # (x=width,y=height)
# 3. plot data 
ax.plot(x,y) 
# 4. Customiz plot 
ax.set(title='Simple Plot', 
      xlabel='x-axis', 
      ylabel='y-axis') 
# 5. Save and show (Save the figure you have coded)
fig.savefig('sample-plot.png')
```

# Two Options for subplots

```python
# Subplot option 1
fig, ((ax0,ax1),(ax2,ax3))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))

# plot to each different axis 
ax0.plot(x, x/2);
ax1.scatter(np.random.random(10), np.random.random(10))
ax2.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax3.hist(np.random.randn(1000));
```

```python
# Subplot option 2 
fig, ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
# plot each different axis 
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

# Which one should we use? (pyplot vs matplotlib OO method?)
* When plotting something quickly, okay to use pyplot method
* when plotting something more advance, use OO method

```python
heart_disease.head()
```

```python
over_50=heart_disease[heart_disease['age']>50]
over_50.head()
```

```python
# pyplot method 
over_50.plot(x='age',y='chol',c='target',kind='scatter');  # we have to denote axes like x,y and c otherwise the code would not run for three axes
# Add title
plt.title('Heart Disease Cholestrol level');
```

```python
# OO method mixed with pyplot
fig, ax=plt.subplots(figsize=(10,6)) 
over_50.plot(x='age',y='chol',c='target',kind='scatter',ax=ax);
```

```python
# OO Method from Scrath
fig, ax=plt.subplots(figsize=(10,6))
# plot the data 
scatter=ax.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])
# customize the plot 
ax.set(title='Heart Disease cholestrol levels', 
      xlabel='Age', ylabel='Cholestrol')
# Add legend
ax.legend(*scatter.legend_elements(),title='Target');
# add a horizontal line 
ax.axhline(over_50['chol'].mean(),linestyle='--');
```

```python
over_50.head()
```

```python
# suplots of chol , age and thalach 
fig, (ax0,ax1)=plt.subplots(nrows=2,ncols=1, figsize=(10,10))
# costumize ax0 
# plot the data 
scatter=ax0.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])
# customize the plot 
ax0.set(title='Heart Disease cholestrol levels', 
       ylabel='Cholestrol')
# Add legend
ax0.legend(*scatter.legend_elements(),title='Target');
# add a horizontal line 
ax0.axhline(over_50['chol'].mean(),linestyle='--')
# Customize ax1
# plot the data 
scatter=ax1.scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target'])
# customize the plot 
ax1.set(title='Heart Disease and max heart rate', 
      xlabel='Age', ylabel='Max Heart rate ')
# Add legend
ax1.legend(*scatter.legend_elements(),title='Target');
# add a horizontal line 
ax1.axhline(over_50['thalach'].mean(),linestyle='--');
# add title to the figure
fig.suptitle('Heart Diseas Analysis')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use( 'seaborn-v0_8-whitegrid')
## OO method from Scratch 
fig, (ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,6))
# Plot the Data 
scatter=ax0.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'],cmap='winter')
# customize the plot 
ax0.set(title='Heart Disease and Cholestrol Levels', 
      ylabel='Cholesterol')
# Change the ax0 x-axis limit
ax0.set_xlim([50,80])
# Add a legend
ax0.legend(*scatter.legend_elements(),title=['Target']);
# Add a horizontal line 
ax0.axhline(over_50['chol'].mean(),linestyle='--',color='red');

scatter=ax1.scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target'],cmap='winter')
# customize the plot 
ax1.set(title='Heart Disease and Max heart rate',
        xlabel='Age', ylabel='Max Heart rate ')
# Change the ax1 axis limit
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
# Add legend
ax1.legend(*scatter.legend_elements(),title='Target');
# add a horizontal line 
ax1.axhline(over_50['thalach'].mean(),linestyle='--',color='red');
# add title to the figure
fig.suptitle('Heart Diseas Analysis',color='black');
# Set red border for ax0
for spine in ax0.spines.values():
    spine.set_edgecolor('red')
# Set red border for ax1
for spine in ax1.spines.values():
    spine.set_edgecolor('red')
# Add space between the plots
plt.subplots_adjust(hspace=0.4)
# Set red border using patch
fig.patch.set_edgecolor('blue')
fig.patch.set_linewidth(2)

# 5. Save and show (Save the figure you have coded)
fig.savefig('Heart Diseas Analysis.png')
```
