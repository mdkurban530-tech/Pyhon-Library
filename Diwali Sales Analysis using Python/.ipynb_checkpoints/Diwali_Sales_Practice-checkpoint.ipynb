```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
df=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
df
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.isnull()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
    df.rename(columns={'Marital_Status':'Shadi'})   # only checking the function how to rename but not fixing the column name
```

```python
df['Amount']=df['Amount'].astype(int)
```

```python
df.describe()
```

```python
df.columns
```

# Analysis using Gender

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers: 
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen,hue = 'Gender')
```

# From the above two graphs we can Analyse that most buyers are females and even money-spending power is also higher for females 

```python
df.columns
```

# Analysis using Age Group

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')
for bars in ax.containers: 
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age,hue='Age Group')
```

# from above graphs most of the Buyers are women between Age group 26-35

```python
df.columns
```

# Analysis Sales Orders and Sales Amount Group by States

```python
sales_orders=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.barplot(x='State',y='Orders',data=sales_orders)
sns.set(rc={'figure.figsize':(15,10)})
```

```python
sales_amount=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(15)
sns.barplot(x='State',y='Amount',data=sales_amount)
sns.set(rc={'figure.figsize':(20,10)})
```

# we can see from above graphs that the most orders and sales are from Up, Maharashtra and karnataka

```python
df.columns
```

# Analysis using Marital Status

```python
ax=sns.countplot(x='Marital_Status',data=df)
sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers: 
    ax.bar_label(bars)
```

```python
AmountbyMaritalStatus=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='Marital_Status',y='Amount',data=AmountbyMaritalStatus,hue='Gender')
```

# from the above graphs we can observe that most of the purchases done by the married women and also the spent the higher amount

```python
df.columns
```

# Analysis using Occupation

```python
ax=sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(20,5)})
for bars in  ax.containers: 
    ax.bar_label(bars)
```

```python
Sales_Occupation=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation',y='Amount',data=Sales_Occupation)
sns.set(rc={'figure.figsize':(20,5)})
```

# Purchasing done by most IT sectors, healthcare and aviation respectively

```python
df.columns
```

# Analysis using Product Category

```python
ax=sns.countplot(x='Product_Category',data=df)
sns.set(rc={'figure.figsize':(20,5)})
```

```python
sales_category=df.groupby(['Product_Category'] , as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)
sns.barplot(x='Product_Category',y='Amount',data=sales_category)
sns.set(rc={'figure.figsize':(20,5)})
```

# We can see that the most sold products are Food,clothing & Apparel and Electronics respectively

```python
df.columns
```
