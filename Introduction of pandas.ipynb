```python
# How to add a column
seats_column=pd. Series([5,5,5,5])
car_sales['Seats']=seats_column.astype(float)
car_sales
```

```python
# add column using python list
fuel_economy=[1.0,2,3,4,5,6,7,8,9,0]          # if we add column using list we have to match the exact same index of our data from
car_sales['Fuel_economy']=fuel_economy
car_sales
```

```python
#add a column using dataframe columns
car_sales['Total_fuel_used (Lit)']=(car_sales['Odometer (KM)']/100) * car_sales['Fuel_economy']
car_sales
```

```python
car_sales['No.of wheels']=4
car_sales
```
